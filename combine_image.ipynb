```python
from PIL import Image
import matplotlib.pyplot as plt

# File paths for the five images
apple_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/fruit/apples/2_100.jpg'  # Update with the actual path
ginger_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/vegetable/Ginger/16_100.jpg'  # Update with the actual path
capsicum_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/vegetable/Capsicum/0492.jpg'  # Update with the actual path
cauliflower_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/vegetable/Cauliflower/0013.jpg'  # Update with the actual path
cucumber_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/vegetable/Cucumber/0520.jpg'  # Update with the actual path

# Load the images
apple_img = Image.open(apple_path).convert('RGB')
ginger_img = Image.open(ginger_path).convert('RGB')
capsicum_img = Image.open(capsicum_path).convert('RGB')
cauliflower_img = Image.open(cauliflower_path).convert('RGB')
cucumber_img = Image.open(cucumber_path).convert('RGB')

# Resize images to the same size for consistency (optional, adjust size as needed)
size = (350, 200)  # Width, Height in pixels
apple_img = apple_img.resize(size)
ginger_img = ginger_img.resize(size)
capsicum_img = capsicum_img.resize(size)
cauliflower_img = cauliflower_img.resize(size)
cucumber_img = cucumber_img.resize(size)

# Create a new blank image with enough width to hold all five images side by side
combined_width = size[0] * 5  # Five images side by side
combined_height = size[1]  # Height remains the same as one image
combined_img = Image.new('RGB', (combined_width, combined_height), 'white')

# Paste the images into the combined image with gaps
gap = 20  # Gap between images in pixels
combined_img.paste(apple_img, (0, 0))  # Apple on the far left
combined_img.paste(ginger_img, (size[0] + gap, 0))  # Ginger with gap
combined_img.paste(capsicum_img, (size[0] * 2 + gap * 2, 0))  # Capsicum with gap
combined_img.paste(cauliflower_img, (size[0] * 3 + gap * 3, 0))  # Cauliflower with gap
combined_img.paste(cucumber_img, (size[0] * 4 + gap * 4, 0))  # Cucumber with gap

# Save the combined image
output_path = 'C:/Users/Manhab Zafar/Desktop/DIP-Project/fruit/abc.jpg'
combined_img.save(output_path)
print(f"Combined image saved at: {output_path}")

# Display the combined image (optional, for verification)
plt.imshow(combined_img)
plt.axis('off')
plt.show()
```
